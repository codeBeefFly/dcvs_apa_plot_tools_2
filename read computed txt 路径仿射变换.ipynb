## rough calculate transform matrix 20210607

````python
import numpy as np



# 单位 米：m

x = 0
y = 0

x_r = 9.58
y_r = 5.52
x_f = 9.58
y_f = 2.69

````

````python
from math import sqrt, cos, sin, radians, atan2, degrees

"""
旋转平移矩阵（先旋转，再平移）

cosa -sina t_x
sina  cosa t_y
0     0    1

"""

# test,angle(30) --> radians
theta_test = radians(30)
print("angle[30] --> radians[%f]" %(theta_test))

# 求相对坐标系对绝对坐标系的旋转角度 theta（弧度）
dx = x_r - x_f  # x方向后轴相对前轴平移量
dy = y_r - y_f  # y方向后轴相对前轴平移量
print("dx = %f, dy = %f" %(dx, dy))

theta = atan2(dy, dx)  # [-pi, pi]
angle = degrees(theta)
print("radians[%f] --> angle[%f]" %(theta, angle))

# 求相对坐标系对绝对坐标系的x，y方向平移量 
t_x = x_r
t_y = y_r

# 旋转平移矩阵
T = np.array([
    [cos(theta), -sin(theta), t_x],
    [sin(theta), cos(theta), t_y],
    [0, 0, 1],
])
print("旋转平移矩阵 T:\n", T)



#############################################################################
x_0 = 0
y_0 = 0
x_1 = 0
y_1 = 1

theta_wc = atan2((y_0 - y_1), (x_0 - x_1))
angle_wc = degrees(theta_wc)
print("radians_wc[%f] --> angle_wc[%f]" %(theta_wc, angle_wc))


theta_cc = atan2(dy, dx)  # [-pi, pi]
angle_cc = degrees(theta_cc)
print("radians_cc[%f] --> angle_cc[%f]" %(theta_cc, angle_cc))

theta = theta_wc - theta_cc
angle = degrees(theta)
print("radians[%f] --> angle[%f]" %(theta, angle))


# 旋转平移矩阵
T_2 = np.array([
    [cos(theta), -sin(theta), t_x],
    [sin(theta), cos(theta), t_y],
    [0, 0, 1],
])
print("旋转平移矩阵 T_2:\n", T_2)
````

````python
# 计算 相对坐标系中点 [0，0] 在绝对坐标系中的位置
p_relative = np.array([
    [0],  # x
    [0],  # y
    [1],  # z
])

p_world = T.dot(p_relative)
print("p_r 在 p_w 的坐标: \n", p_world)
````
