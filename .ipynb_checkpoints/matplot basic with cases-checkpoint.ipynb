## 1. draw a single graph

```python
# draw a single graph
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

## 2. draw single graph with [x, y]

```python
# draw single graph with [x, y]
x_list = [0, 1, 3, 5, 7]
y_list = [0, 2, 4, 6, 8]

plt.plot(x_list, y_list)
```

## 3. draw points

```python
# draw single graph with [x, y]
x_list = [0, 1, 3, 5, 7]
y_list = [0, 2, 4, 6, 8]

#创建图并命名
plt.figure('Scatter fig')
ax = plt.gca()  # get current axes

#设置x轴、y轴名称
ax.set_xlabel('x --->')
ax.set_ylabel('y --->')

#画散点图，以x_list中的值为横坐标，以y_list中的值为纵坐标
#参数c指定点的颜色，s指定点的大小,alpha指定点的透明度
ax.scatter(x_list, y_list, c='r', s=20, alpha=0.5)

plt.title("Scatter fig")
plt.show()
```

## 4. draw line segment

```python
# draw single graph with [x, y]
x_list = [0, 1, 3, 5, 7]
y_list = [0, 2, 4, 6, 8]

#创建图并命名
plt.figure("line segment")
ax = plt.gca()  # get current axes

#设置x轴、y轴名称
ax.set_xlabel('x')
ax.set_ylabel('y')

#画散点图，以x_list中的值为横坐标，以y_list中的值为纵坐标
#参数c指定点的颜色，s指定点的大小,alpha指定点的透明度
ax.scatter(x_list, y_list, c='r', s=30, alpha=0.5)

#画连线图，以x_list中的值为横坐标，以y_list中的值为纵坐标
#参数c指定连线的颜色，linewidth指定连线宽度，alpha指定连线的透明度
ax.plot(x_list, y_list, color='g', linewidth=1, alpha=0.6)

plt.title("line segment")
plt.show()
```

## 5. enable gridding

```python
# draw single graph with [x, y]
x_list = [0, 1, 3, 5, 7]
y_list = [0, 2, 4, 6, 8]

#创建图并命名
plt.figure("line segment")
ax = plt.gca()  # get current axes

#设置x轴、y轴名称
ax.set_xlabel('x --->')
ax.set_ylabel('y --->')

#画散点图，以x_list中的值为横坐标，以y_list中的值为纵坐标
#参数c指定点的颜色，s指定点的大小,alpha指定点的透明度
ax.scatter(x_list, y_list, c='r', s=30, alpha=0.5)

#画连线图，以x_list中的值为横坐标，以y_list中的值为纵坐标
#参数c指定连线的颜色，linewidth指定连线宽度，alpha指定连线的透明度
ax.plot(x_list, y_list, color='g', linewidth=1, alpha=0.6)

plt.title("line segment")
plt.grid(linestyle='-.')  # 生成网格
plt.show()
```
